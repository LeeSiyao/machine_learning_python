```python
class linearRegression:
    """线性回归（梯度下降方法）。
        y=w0+w1x1+w2x2+w3x3+..."""
    
    def _init_(self,alpha,times):
        """初始化方法。
        parameters
        -----
        alpha : float
            学习率，用于控制步长（权重调整的幅度）
        times : int
            循环迭代次数（调整权重值的次数）
        """
        self.alpha = alpha
        self.times = times
        
    def fit(self,X,y):
        """根据提供的训练数据，对模型进行训练
        parameters：
        -----
        X : 类数组类型。形状：[样本数量，特征数量]
            待训练的样本特征属性。（特征矩阵）
        y : 类数组类型。形状：[样本数量]
            目标值（标签信息）。
        """
        X = np.asarray(X)
        y = np.asarray(y)
        # 创建权重向量，初始值为0（或任何其他的值），长度比特征数量多1（多出一个截距 ）。
        self.w_ = np.zeros(1 + X.shape[1]) #.shape返回矩阵行列数，shape[1]指向样本特征的数量。
        # 创建损失列表，用于保存每次迭代后的损失值。损失值计算：（损失值-真实值）的平方和/2
        self.loss_ = []
        
        # 进行循环，在每次迭代过程中不断调整权重值，使损失值不断减少。
        for i in range(self.times):
            # 计算预测值
            y_hat = np.dot(self.w_[1:],X) + self.w_[0]
            # 计算预测值和真实值的差距
            error = y - y_hat
            # 将损失值加入到损失列表中
            self.loss_.append(np.sum(error ** 2) / 2)
            # 根据差距调整权重w_,根据公式：调整为 权重（j）=权重（j）+学习率*sum（（y-y_hat）*x（j））
            self.w_[0] += self.alpha * np.sum(error)
            self.w_[1:] += self.alpha * np.dot(X.T,error) # 对X矩阵进行转置，使每一个特征列变成行，以此跟标签矩阵相乘
            
    def predict(self,X):
        """根据传递参数的样本，对样本数据进行预测。
        parameter
        -----
        X：类数组类型，形状[样本数量，特征数量]
           待测试样本。
        
        returns
        -----
        result：数组类型。
            预测的结果。
        """
        X = np.asarray(X)
        result = np.dot(X,self.w_[1:]) + self.w_[0]
        return result
            
```
